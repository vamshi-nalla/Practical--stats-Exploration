# Estimates of Variability


---
When we summarize a feature in data, there are two important aspects to consider. The first is **location** — this tells us about the central value of the data (like the average). The second is **variability**, also known as **dispersion** — this tells us how spread out or clustered the data values are.

In statistics, **variability** is extremely important. That’s because:

* We need to know how to **measure** it,
* How to **reduce** it when necessary,
* How to **tell the difference** between random variation and real, meaningful differences,
* How to **identify the causes** behind real variation,
* And how to **make good decisions** even when variation is present.

So, in short: statistics is largely about understanding and working with **variability**.

---

# Key Terms for Variability Metrics or Measures of Variability

### Deviations

> The difference between the observed values and the estimate of location (like the mean or median).
* **Synonyms:** `errors`, `residuals`


### Variance
> The sum of **squared deviations** from the mean, divided by *(n – 1)*, where *n* is the number of data values.
* **Synonym:** `mean-squared-error`


### Standard Deviation

> The **square root** of the variance.
* Measures average spread around the mean in the original units.



### Mean Absolute Deviation (MAD)

> The **mean** of the **absolute values** of the deviations from the mean.
* **Synonyms:** `l1-norm`, `Manhattan norm`



### Median Absolute Deviation from the Median (MAD-M)

> The **median** of the absolute values of the deviations from the **median**.
* More robust to outliers than standard deviation or mean absolute deviation.


### Range

> The difference between the **largest** and **smallest** value in a data set.


### Order Statistics

> Metrics based on data values sorted from **smallest to largest**.
* **Synonym:** `ranks`


###  Percentile

> The value below which **P percent** of the data falls.
* Example: The 90th percentile is the value below which 90% of the data lies.
* **Synonym:** `quantile`


### Interquartile Range (IQR)

> The difference between the **75th percentile (Q3)** and the **25th percentile (Q1)**.
* **Synonym:** `IQR`

## Why Do We Use **n – 1** in Variance?

### — Understanding **Degrees of Freedom**

When calculating the **variance** of a **sample**, we often divide by `n – 1` instead of just `n`. This is a common question in statistics. Here’s the simple explanation:

###  The Problem

Let’s say you want to estimate how spread out a population is — but you only have a **sample** of data (not the full population).

* If you divide by `n` when computing sample variance, the result is usually **too small**. That’s because it **underestimates** the actual population variance.
* This is called a **biased estimate**.

To correct this, we divide by `n – 1` — which gives us an **unbiased estimate** of the population variance.



### Why `n – 1`?

This has to do with a concept called **degrees of freedom**.

* When you calculate the **sample mean**, you use all `n` data points.
* Once the sample mean is fixed, **only `n – 1` of those data points can vary freely**. The last value is **constrained** (it must "fit" the mean).
* Because of that one constraint, we say there are only **`n – 1` degrees of freedom** left for estimating variance.

So, we divide by `n – 1` to account for that lost freedom.

###  Final Notes

* In **large samples**, the difference between dividing by `n` or `n – 1` becomes very small.
* But in **small samples**, dividing by `n – 1` is important to get an accurate (unbiased) estimate.
* This whole idea is part of the concept called **degrees of freedom** — how many data points are truly "free" to vary.

## Are Variance, Standard Deviation, MAD, and Median MAD the Same?

**No**, these are **not equivalent**, even when the data follows a **normal distribution**. Let's break it down.



### Four Common Measures of Variability:

1. **Variance** — Average of squared deviations from the mean
2. **Standard Deviation (SD)** — Square root of the variance
3. **Mean Absolute Deviation (MAD)** — Average of absolute deviations from the mean
4. **Median Absolute Deviation from the Median (Median MAD)** — Median of absolute deviations from the median


### These Measures Are Not Equal

Even if your data comes from a **normal distribution**, the values of these metrics are **not the same**.
Standard Deviation > Mean Absolute Deviation > Median Absolute Deviation

* **Standard deviation** is always the largest of the three
* **Mean absolute deviation** is smaller than SD
* **Median absolute deviation** is typically the smallest


### Why Use a Scaling Factor?

To **compare** the Median Absolute Deviation (MAD from median) with the **Standard Deviation**, we often **multiply it by a constant**.
* Common scaling factor: **1.4826**

This factor is chosen because:
> In a **normal distribution**, about **50%** of the values fall within 1.4826 times of (MAD from median)
* This makes the scaled MAD behave similarly to the standard deviation **in the context of normal data**.

### Summary
* These measures all describe **how spread out** data is.
* They differ in how they treat **outliers** and what they measure around (mean vs median).
* **Standard deviation** is sensitive to outliers.
* **Median MAD** is more **robust** (less affected by extreme values).
* To make **Median MAD** comparable to **Standard Deviation**, multiply it by **1.4826** if the data is normally distributed.

### Estimates Based on Percentiles

One way to measure how spread out data is involves looking at the sorted data. When we organize data from smallest to largest, the statistics we get from this order are called **order statistics**.

The simplest measure of spread is the **range**—the difference between the largest and smallest values. While knowing the minimum and maximum is useful, the range is very sensitive to **outliers** (values that are much higher or lower than most of the data), so it’s not the best way to understand variability.

To reduce the effect of outliers, we look at the spread after ignoring some values at the low and high ends. This is done by calculating differences between **percentiles**.

A **percentile** is a value below which a certain percentage of the data falls. For example, the 80th percentile is the value below which 80% of the data lies. The **median** is just the 50th percentile, meaning half the data is below it and half above it. Percentiles are similar to **quantiles**, just expressed as percentages instead of fractions.

A common way to measure variability is the **interquartile range (IQR)**, which is the difference between the 75th percentile (Q3) and the 25th percentile (Q1). For example, take this data set: {3, 1, 5, 3, 6, 7, 2, 9}. Sorting it gives {1, 2, 3, 3, 5, 6, 7, 9}. The 25th percentile is about 2.5, and the 75th percentile is about 6.5. So, the IQR = 6.5 - 2.5 = 4.

When the data set is very large, calculating exact percentiles by sorting all the data can be slow. So, special algorithms are used in machine learning and statistics software to quickly estimate percentiles with guaranteed accuracy.
